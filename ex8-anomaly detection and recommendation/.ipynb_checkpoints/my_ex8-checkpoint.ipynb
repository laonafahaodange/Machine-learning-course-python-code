## 异常检测（anomaly detection）
使用高斯模型检测一个二维数据集（未标记）的示例是否为异常

```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
```

```python
#读取数据
data=loadmat('data/ex8data1.mat')
X=data['X']
Xval=data['Xval']
yval=data['yval']
X.shape,Xval.shape,yval.shape
```

```python
#可视化
fig,ax=plt.subplots(figsize=(12,8))
ax.scatter(X[:,0],X[:,1],s=10)
plt.show()
```

接下来需要根据每个特征估计高斯分布，因此定义一个函数求得每个特征的特征值和方差来构造每个特征的高斯分布

<img style="float: left;" src="../img/Gaussian distribution.png">

```python
def estimateGaussian(X):
    u=X.mean(axis=0)#均值
    sigma_square=X.var(axis=0)
    
    return u,sigma_square
```

```python
u,sigma_square=estimateGaussian(X)
u,sigma_square
```

```python
#scipy内置正态分布
from scipy import stats
norm_dist_x1=stats.norm(u[0],sigma_square[0])
norm_dist_x2=stats.norm(u[1],sigma_square[1])
#测试下u约等于14时对应的概率密度
norm_dist_x1.pdf(14)#pdf=probability density function
```

<img style="float: left;" src="../img/pdf.png">

```python
#计算数据集中每个样本在高斯分布上对应的概率密度
p=np.zeros_like(X)
p[:,0]=norm_dist_x1.pdf(X[:,0])
p[:,1]=norm_dist_x2.pdf(X[:,1])
p.shape
```

```python
#同样为交叉验证集中的样本做同样操作
pval=np.zeros_like(Xval)
pval[:,0]=norm_dist_x1.pdf(Xval[:,0])
pval[:,1]=norm_dist_x2.pdf(Xval[:,1])
pval.shape
```

<img style="float: left;" src="../img/evaluate.png">
接下来，我们建立评价体系，根据交叉验证集选择合适的epsilon，参考代码给的方法是从范围pval_min到pval_max按照一定步长遍历epsilon，再根据当前epsilon计算F1score，根据得分最高选出epsilon

```python
def selectThreshold(pval,yval):
    best_epsilon=0
    best_f1_score=0
    f1_score=0
    
    preds_final=np.zeros((pval.shape[0],1))
    
    step=(pval.max()-pval.min())/1000
    
    #range不支持步长为小数，而np.arange支持步长为小数
    for epsilon in np.arange(pval.min(),pval.max(),step):
        #标签y=1为异常
        preds=pval<epsilon
        preds_final[:,0]=preds[:,0]|preds[:,1]
        
        #这里我觉得参考代码有问题，而且和参考代码理解有些不同，见分析
        tp=np.sum(np.logical_and(preds_final==1,yval==1)).astype(float)
        fp=np.sum(np.logical_and(preds_final==1,yval==0)).astype(float)
        fn=np.sum(np.logical_and(preds_final==0,yval==1)).astype(float)
        
        precision=tp/(tp+fp)
        recall=tp/(tp+fn)
        f1_score=(2*precision*recall)/(precision+recall)
        
        if f1_score>best_f1_score:
            best_f1_score=f1_score
            best_epsilon=epsilon
        
    return best_epsilon,best_f1_score
```

按照理解，preds中的每个元素应该有两个值a和b，取值为True和False之一，按照对异常点的理解可以有很多种，比如当ab都为True时说明当前点一定是异常点（因为都小于epsilon，理解成一个小概率），但是当ab当中仅有一个为True时，就有多种理解了，可以认为是异常也可以认为是正常，这里我当作异常来处理。  
参考代码的问题是，在我调试的过程当中，比如当epsilon=pval.min()+step时，preds只有倒数第二、四、五行出现了True（见下方代码），那么按我的理解来说，异常样本应该有3个，可是实际执行np.sum(np.logical_and(preds == 1, yval == 1))后的结果是5，说明它把出现过的True累加了一遍，说明是不对的。

```python
step=(pval.max()-pval.min())/1000
epsilon=pval.min()+step
preds=pval<epsilon
preds

# preds_new=np.zeros((Xval.shape[0],1))
# preds_new[:,0]=preds[:,0]|preds[:,1]

# np.logical_and(preds == 1, yval == 1)
# np.sum(np.logical_and(preds_new == 1, yval == 1))
```

```python
epsilon,f1_score=selectThreshold(pval,yval)
epsilon,f1_score
```

```python
#我们可视化上述参数下的这些异常点
anomaly_points_final=np.zeros((p.shape[0],1))
anomaly_points=p<epsilon
anomaly_points_final[:,0]=anomaly_points[:,0]|anomaly_points[:,1]
anomaly_points_idx=np.where(anomaly_points_final)
anomaly_points_idx=anomaly_points_idx[0]
anomaly_points_idx
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.scatter(X[:,0],X[:,1])
ax.scatter(X[anomaly_points_idx,0],X[anomaly_points_idx,1],c='r',marker='o')
plt.show()
```

## 协同过滤

接下来定义协同过滤的代价函数
<img style="float: left;" src="../img/rcmd_cost.png">

```python
def cost(params,Y,R,features_num):
    movie_num=Y.shape[0]
    user_num=Y.shape[1]
    
    Y=np.matrix(Y)#1682 x 943
    R=np.matrix(R)#1682 x 943
    
    #重新拆开参数，还原X和theta
    X=params[:movie_num*features_num]
    X=np.reshape(X,(movie_num,features_num))#1682 x 10
    X=np.matrix(X)
    theta=params[movie_num*features_num:]
    theta=np.reshape(theta,(user_num,features_num))#943 x 10
    theta=np.matrix(theta)
    
    #multiply是对应元素相乘，作用相当于公式里面的r(i,j)=1
    error=np.multiply((X*theta.T-Y),R)#1682 x 943
    error_square=np.power(error,2)
    J=1/2*np.sum(error_square)
    
    return J
```

```python
#预训练参数，包含我们需要的theta和X
data=loadmat('data/ex8_movieParams.mat')
X=data['X']
theta=data['Theta']
X.shape,theta.shape
```

```python
#取一小部分来测试
users=4
movies=5
features=3

X_sub=X[:movies,:features]#5 x 3
theta_sub=theta[:users,:features]#4 x 3
Y_sub=Y[:movies,:users]#5 x 4
R_sub=R[:movies,:users]#5 x 4

#5 x 3 + 4 x 3 ，相当于把参数X和theta组装一下作为参数传给cost函数
params=np.concatenate((np.ravel(X_sub),np.ravel(theta_sub)))
cost(params,Y_sub,R_sub,features)
```

根据公式优化下cost函数，使其能返回梯度值
<img style="float: left;" src="../img/collaborative filtering2.png">
<img style="float: left;" src="../img/collaborative filtering3.png">

```python
def cost(params,Y,R,features_num,learning_rate):
    movie_num=Y.shape[0]
    user_num=Y.shape[1]
    
    Y=np.matrix(Y)#1682 x 943
    R=np.matrix(R)#1682 x 943
    
    #重新拆开参数，还原X和theta
    X=params[:movie_num*features_num]
    X=np.reshape(X,(movie_num,features_num))#1682 x 10
    X=np.matrix(X)
    theta=params[movie_num*features_num:]
    theta=np.reshape(theta,(user_num,features_num))#943 x 10
    theta=np.matrix(theta)
    
    #multiply是对应元素相乘，作用相当于公式里面的r(i,j)=1
    error=np.multiply((X*theta.T-Y),R)#1682 x 943
    error_square=np.power(error,2)
    first_term=1/2*np.sum(error_square)
    second_term=learning_rate/2*(np.sum(np.power(X,2)))
    third_term=learning_rate/2*np.sum((np.power(theta,2)))
    J=first_term+second_term+third_term
    
    x_grad=error*theta+learning_rate*X
    theta_grad=error.T*X+learning_rate*theta
    
    grad=np.concatenate((np.ravel(x_grad),np.ravel(theta_grad)))
    
    return J,grad
```

```python
J,grad=cost(params,Y_sub,R_sub,features,1.5)
J,grad
```

接下来通过读取文件，根据数据集中的索引对应其表示的电影名称

```python
#通过字典的方式来保存
movie_idx={}
f=open('data/movie_ids.txt',encoding='gbk')
for line in f:
    tokens=line.split(' ',1)#只根据第一个空格进行分割
    movie_idx[int(tokens[0])-1]=tokens[1][:-1]#-1是为了去掉换行符\n
#测试
movie_idx[0]
```

接下来我们使用练习中提供的评分，本意是让我们对一些电影做出评分，然后训练算法会根据我们的评分为我们推荐电影

```python
ratings=np.zeros((1682,1))

ratings[0]=4
ratings[6]=3
ratings[11] = 5
ratings[53] = 4
ratings[63] = 5
ratings[65] = 3
ratings[68] = 5
ratings[97] = 2
ratings[182] = 4
ratings[225] = 5
ratings[354] = 5

r=np.zeros((1682,1))

r[0]=1
r[6]=1
r[11] = 1
r[53] = 1
r[63] = 1
r[65] = 1
r[68] = 1
r[97] = 1
r[182] = 1
r[225] = 1
r[354] = 1

print('Rated {0} with {1} stars.'.format(movie_idx[0], str(int(ratings[0]))))
print('Rated {0} with {1} stars.'.format(movie_idx[6], str(int(ratings[6]))))
print('Rated {0} with {1} stars.'.format(movie_idx[11], str(int(ratings[11]))))
print('Rated {0} with {1} stars.'.format(movie_idx[53], str(int(ratings[53]))))
print('Rated {0} with {1} stars.'.format(movie_idx[63], str(int(ratings[63]))))
print('Rated {0} with {1} stars.'.format(movie_idx[65], str(int(ratings[65]))))
print('Rated {0} with {1} stars.'.format(movie_idx[68], str(int(ratings[68]))))
print('Rated {0} with {1} stars.'.format(movie_idx[97], str(int(ratings[97]))))
print('Rated {0} with {1} stars.'.format(movie_idx[182], str(int(ratings[182]))))
print('Rated {0} with {1} stars.'.format(movie_idx[225], str(int(ratings[225]))))
print('Rated {0} with {1} stars.'.format(movie_idx[354], str(int(ratings[354]))))
```

将自己的评价向量加入现有数据集中(理解成新增加一个用户)

```python
data=loadmat('data/ex8_movies.mat')
Y=data['Y']
Y=np.append(Y,ratings,axis=1)
Y.shape
```

**注：函数np.append(arr, values, axis=None)，为原始array添加一些values**

```python
R=data['R']
R=np.append(R,r,axis=1)
R.shape
```

先进行数据的初始化和归一化

```python
movies=Y.shape[0]#1682
users=Y.shape[1]#944
features=10
learning_rate=10

#随机浮点型参数，范围0-1，大小为参数size
X=np.random.random(size=(movies,features))
theta=np.random.random(size=(users,features))
params=np.concatenate((np.ravel(X),np.ravel(theta)))
```

```python
#归一化，即每部电影的评分减去每部电影的平均分
Y_norm=np.zeros_like(Y)
Y_mean=np.zeros((movies,1))

for i in range(movies):
    idx=np.where(R[i,:]==1)[0]#评过分的用户索引
    mean=np.sum(Y[i,:])/len(idx)
    Y_mean[i]=mean
    Y_norm[i,idx]=Y[i,idx]-mean

#归一化后，Y_norm的均值理论上应该等于0
Y_norm.mean()
```

```python
#开始激动人心的训练
from scipy.optimize import minimize

fmin=minimize(fun=cost,x0=params,args=(Y_norm,R,features,learning_rate),
              method='CG',jac=True,options={'maxiter':100})
fmin
```

求最小化代价函数对应的参数  
fun：该参数就是costFunction，要去最小化的损失函数，损失函数在定义时，theta必须为第一个参数且其shape必须为(n,)即一维数组(终于知道为啥参数要打包成params了...)  
x0：初始化的theta,其shape必须为shape(n,)即一维数组  
args：其他参数，如X，Y，lambda等
method：该参数代表采用的方式，默认是BFGS, L-BFGS-B, SLSQP中的一种，可选TNC  
jac：该参数就是计算梯度的函数，和fun参数类似，第一个必须为theta且其shape必须为(n,)即一维数组,最后返回的梯度也必须为一个一维数组。如果jac是布尔值且为真，则假定fun返回objective和gradient作为（f，g）元组  
options：迭代次数  

```python
#还原最小代价函数下的X和theta
X=fmin.x[:movies*features]
X=np.reshape(X,(movies,features))#1682 x 10
X=np.matrix(X)
theta=fmin.x[movies*features:]
theta=np.reshape(theta,(users,features))#944 x 10
theta=np.matrix(theta)

X.shape,theta.shape
```

```python
#预测评分
predictions=X*theta.T#1682 x 944
my_preds=predictions[:,-1]+Y_mean#最后一列就是预测我们的分数
my_preds.shape
```

```python
my_preds=np.array(my_preds)
idx=np.argsort(my_preds,axis=0)[::-1]#argsort从小到大排序，所以需要[::-1]逆序
idx
```

```python
print('top 10 movie recommendation:\n')
for i in range(10):
    j=int(idx[i])
    print('Predicting rating {} for movie {}'.format(my_preds[j],movie_idx[j]))
```
