```python
#定义sigmoid函数
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#可视化sigmoid函数
test_x=np.arange(-10,10,step=0.1)
test_y=sigmoid(test_x)

fig,ax=plt.subplots(figsize=(12,8))
ax.plot(test_x,test_y,'g')

plt.show()
```

编写代价函数：
$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}$

```python
def cost(theta,X,y):
    #矩阵化
    theta=np.matrix(theta)
    X=np.matrix(X)
    y=np.matrix(y)
    #式子比较长，拆开吧
    h_theta=sigmoid(X*theta.T)
    first=np.multiply(-y,np.log(h_theta))
    second=np.multiply((1-y),np.log(1-h_theta))
    J_val=np.sum(first-second)/(len(X))
    return J_val
```

```python
# 插入第一列，表头名称，插入数字
data.insert(0,'Ones',1)

#检查维度
cols=data.shape[1]#列数cols分别对应：Ones Exam1 Exam2 Admitted
X=data.iloc[:,0:cols-1]
y=data.iloc[:,cols-1:cols]

#格式转换
X=np.array(X.values)
y=np.array(y.values)

#theta初始化
theta=np.zeros(3)
```

# gradient descent(梯度下降)
* 这是批量梯度下降（batch gradient descent）  

$$\frac{\partial J\left( \theta  \right)}{\partial {{\theta }_{j}}}=\frac{1}{m}\sum\limits_{i=1}^{m}{({{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}})x_{_{j}}^{(i)}}$$
* 转化为向量化计算： $\frac{1}{m} X^T( Sigmoid(X\theta) - y )$

```python
#仅执行一次梯度下降
def gradient(theta,X,y):
    theta=np.matrix(theta)
    X=np.matrix(X)
    y=np.matrix(y)
    
    parameters=int(theta.ravel().shape[1])#这里的theta已经矩阵化了，是1*3的矩阵
    
    grad=np.zeros(parameters)
    
    error=sigmoid(X*theta.T)-y
    
    for i in range(parameters):
        term=np.multiply(error,X[:,i])#这里的i应该对应公式里的j，公式里的i表示行j表示列
        grad[i]=np.sum(term)/(len(X))
        
    return grad
```

```python
gradient(theta,X,y)
```

PS:可以用SciPy's truncated newton（TNC）实现寻找最优参数，截断牛顿法

```python
#用scipy需要最优参数的例子
import scipy.optimize as opt
result=opt.fmin_tnc(func=cost,x0=theta,fprime=gradient,args=(X,y))
result
```

参数：  
func：优化的目标函数  
x0：初值  
fprime：提供优化函数func的梯度函数，不然优化函数func必须返回函数值和梯度，或者设置approx_grad=True  
approx_grad :如果设置为True，会给出近似梯度  
args：元组，是传递给优化函数的参数  

```python
#该theta下的代价函数结果
cost(result[0],X,y)
```

逻辑回归模型的假设函数： 
	\\[{{h}_{\theta }}\left( x \right)=\frac{1}{1+{{e}^{-{{\theta }^{T}}X}}}\\] 
当${{h}_{\theta }}$大于等于0.5时，预测 y=1

当${{h}_{\theta }}$小于0.5时，预测 y=0 。

```python
#根据逻辑归回模型的假设函数，尝试利用上面优化算法获得的theta来预测录取概率
def predict(theta,X):
    probability=sigmoid(X*theta.T)
    probability_list=[]
#     return [1 if x >= 0.5 else 0 for x in probability]
    for x in probability:
        if x>=0.5:
            probability_list.append(1)
        else:
            probability_list.append(0)
    return probability_list
```

```python
theta_min=np.matrix(result[0])
predictions=predict(theta_min,X)

correct=0
for i in range(len(predictions)):
    if predictions[i]==y[i]:
        correct=correct+1

accuracy=correct/(len(predictions))
print('accuracy = {0}%'.format(accuracy*100))
```

# regularized cost（正则化代价函数）
$$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}+\frac{\lambda }{2m}\sum\limits_{j=1}^{n}{\theta _{j}^{2}}$$

```python
#超参数learning_rate：学习率，控制正则化项，相当于公式中的lambda
def costReg(theta,X,y,learning_rate):
    theta=np.matrix(theta)
    X=np.matrix(X)
    y=np.matrix(y)
    
    first=np.multiply(-y,np.log(sigmoid(X*theta.T)))
    second=np.multiply((1-y),np.log(1-sigmoid(X*theta.T)))
    reg=learning_rate/(2*len(X))*np.sum(np.power(theta[:,1:theta.shape[1]],2))
    
    return np.sum(first-second)/(len(X))+reg
```

使用梯度下降法令这个代价函数最小化，因为我们未对${{\theta }_{0}}$ 进行正则化，所以梯度下降算法将分两种情形：
\begin{align}
  & Repeat\text{ }until\text{ }convergence\text{ }\!\!\{\!\!\text{ } \\ 
 & \text{     }{{\theta }_{0}}:={{\theta }_{0}}-a\frac{1}{m}\sum\limits_{i=1}^{m}{[{{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}}]x_{_{0}}^{(i)}} \\ 
 & \text{     }{{\theta }_{j}}:={{\theta }_{j}}-a[\frac{1}{m}\sum\limits_{i=1}^{m}{[{{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}}]x_{j}^{(i)}}+\frac{\lambda }{m}{{\theta }_{j}}] \\ 
 & \text{          }\!\!\}\!\!\text{ } \\ 
 & Repeat \\ 
\end{align}

对上面的算法中 j=1,2,...,n 时的更新式子进行调整可得： 
${{\theta }_{j}}:={{\theta }_{j}}(1-a\frac{\lambda }{m})-a\frac{1}{m}\sum\limits_{i=1}^{m}{({{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}})x_{j}^{(i)}}$

```python
#不能理解这里的learning_rate对应公式的什么参数...应该是lambda
#但是按照下面这个函数写法的话，公式里的a应该要去掉，先理解成去掉a吧
def gradientReg(theta,X,y,learning_rate):
    theta=np.matrix(theta)
    X=np.matrix(X)
    y=np.matrix(y)
    
    parameters=theta.ravel().shape[1] #ravel将多维数组转为一维数组
    grad=np.zeros(parameters)
    
    error=sigmoid(X*theta.T)-y
    
    for i in range(parameters):
        term=np.multiply(error,X[:,i])
        if i==0:
            grad[i]=np.sum(term)/len(X)
        else:
            grad[i]=np.sum(term)/len(X)+(learning_rate/len(X))*theta[:,i]
    
    return grad
```

还可以直接调用scikit-learn的包  
penalty: 用于选择正则化项。参数值为‘l1’时，表示正则化项为l1正则化；参数项为’l2‘时表示正则化项为l2正则化。正则化时防止模型过拟合最为常用的手段之一。  
C: 越小的数值表示越强的正则化。  

L1正则化和L2正则化：
L1正则化就是在loss function后边所加正则项为L1范数，加上L1范数容易得到稀疏解（0比较多）。L2正则化就是loss function后边所加正则项为L2范数的平方，加上L2正则相比于L1正则来说，得到的解比较平滑（不是稀疏），但是同样能够保证解中接近于0（但不是等于0，所以相对平滑）的维度比较多，降低模型的复杂度。

```python
from sklearn import linear_model
model=linear_model.LogisticRegression(penalty='l2',C=1.0)#L2正则化，正则化参数C=1
model.fit(X2,y2.ravel())
```

```python
model.score(X2,y2)
```
